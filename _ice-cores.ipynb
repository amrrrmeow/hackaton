```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv(r"\Users\vn_al\OneDrive\Рабочий стол\Elbrus_annual_summer.txt", skipinitialspace=True, delim_whitespace=True, na_values='nan')
data.columns = ["Year", "Nh4", "No3", "So4"]
data.dropna(inplace=True)
data["Year"] = pd.to_numeric(data["Year"])

data

pre_industrial_data = data[(data["Year"] >= 1774) & (data["Year"] <= 1850)]
industrial_data = data[(data["Year"] >= 1989) & (data["Year"] <= 2009)]

pre_industrial_means = pre_industrial_data[["Nh4", "No3", "So4"]].mean()
industrial_means = industrial_data[["Nh4", "No3", "So4"]].mean()

print("Средние показатели за доиндустриальный период (1774-1850):")
print(pre_industrial_means)
print("\nСредние показатели за период 1989-2009 гг.:")
print(industrial_means)

increase_ratio = industrial_means / pre_industrial_means

print("Отношение средних показателей за период 1989-2009 к средним показателям за доиндустриальный период:")
print(increase_ratio)

plt.figure(figsize=(15, 6))

plt.plot(data["Year"], data["Nh4"], label="Аммоний")
plt.plot(data["Year"], data["No3"], label="Нитраты")
plt.plot(data["Year"], data["So4"], label="Сульфаты")

plt.title("Концентрация основных ионов по годам")
plt.xlabel("Год")
plt.ylabel("Концентрация (ppb)")
plt.legend()
plt.grid(True)

plt.xticks(range(min(data["Year"]), max(data["Year"]), 10))

plt.show()





```
